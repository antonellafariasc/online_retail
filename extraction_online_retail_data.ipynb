Validar las inconsistencias en los Datos (UnitPrice) y (Quantity)

```python
# Crear un dataset para identificar inconsistencias de los datos unitprice y quantity
# Filtrar las filas donde la cantidad es negativa
df_inconsistencies = df_online_retail_data[(df_online_retail_data['Quantity'] <= 0) | (df_online_retail_data['UnitPrice'] <= 0)]
df_inconsistencies
```

```python
# Visualizar si hay cantidades en cero
df_qty_inconsistencies = df_online_retail_data[df_online_retail_data['Quantity'] == 0]
df_qty_inconsistencies
```

```python
df_qty_inconsistencies.shape
```

Las cantidades Negativas no son inconistentes, significan que son devoluciones

```python
# Visualizar las inconsistencias en los datos Unit Prce para identificar valores negativos
df_price_inconsistencies_1 = df_online_retail_data[(df_online_retail_data['UnitPrice'] <0)]
df_price_inconsistencies_1
```

Se puede considerar que estos datos no son de ventas de productos.  Estos datos se pueden eliminar.

```python
# Visualizar las inconsistencias en los datos Unit Price para identificar valores cero
df_price_inconsistencies_2 = df_online_retail_data[(df_online_retail_data['UnitPrice'] ==0)]
df_price_inconsistencies_2
```

```python
df_price_inconsistencies_2.shape
```

```python
# Vamos a comprobar que los precios en cero es un error o esta relacionado con que el customerID es un dato null
df_inconsistencies_3 = df_online_retail_data[(df_online_retail_data['UnitPrice'] == 0) & (df_online_retail_data['CustomerID'].isnull())]
df_inconsistencies_3
```

Revisión General de los Datos para validar si existen datos faltantes y calidad de los dato

```python
# Visualizar solo las columnas con datos nulos
print(df_online_retail_data.columns[df_online_retail_data.isnull().any()])
print()
# Visualizar el número de valores nulos por columna
df_online_retail_data.isnull().sum().sort_values(ascending=False)
print("Count of null values per column:\n", df_online_retail_data.isnull().sum().sort_values(ascending=False))
print()
# Visualizar el porcentaje de valores nulos por columna
print('%_null_value:\n', df_online_retail_data.isnull().mean() * 100)
```

Hacer un Breve Análisis de los Datos para obtener información adicional para hacer el Diccionario de Datos

```python
# Validar valores unicos en las columnas
print("Unique values in 'InvoiceNo':", df_online_retail_data['InvoiceNo'].nunique())

print("Unique values in 'StockCode':", df_online_retail_data['StockCode'].nunique())

print("Unique values in 'Country':", df_online_retail_data['Country'].nunique())

print("Unique values in 'CustomerID':", df_online_retail_data['CustomerID'].nunique())
```
