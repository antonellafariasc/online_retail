```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df_online_retail_data = pd.read_csv('online_retail.csv')
df_online_retail_data
```

Completar los datos nulos de la Serie Description

```python
# Crear una copia del DataFrame
df_online_retail_data_filldesc = df_online_retail_data.copy()
```

```python
# Rellenar los valores nulos en la columna 'Description' con valores de la columna 'StockCode'
df_online_retail_data_filldesc["Description"] = df_online_retail_data_filldesc.groupby("StockCode")["Description"].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Descripción desconocida"))
```

```python
df_online_retail_data_filldesc.info()
```

```python
# Verificar 10 valores nulos del df original
df_online_retail_data[df_online_retail_data["Description"].isnull()].head(10)
```

```python
# Verificar 10 valores nulos del df modificado con iloc
print(df_online_retail_data_filldesc.iloc[1970:1988])
```

```python
print(df_online_retail_data_filldesc.iloc[2024:2027])
```

```python
# Crear un DataFrame con las filas donde la descripción es "Descripción desconocida"
df_descripcion_desconocida = df_online_retail_data_filldesc[df_online_retail_data_filldesc["Description"] == "Descripción desconocida"]
```

```python
df_descripcion_desconocida.shape
```

Nota: Una vez rellenado los datos nulos, se tienen ahora datos con "Descripcion Desconocida".  Son 112 datos que por su tamano se considera que no son influyentes.  En tal sentido, Se eliminaran de DF.

```python
# Eliminar los datos con "Descripción desconocida" del DataFrame df_online_retail_data_filldesc
df_online_retail_data_filldesc.drop(df_online_retail_data_filldesc[df_online_retail_data_filldesc['Description'] == 'Descripción desconocida'].index, inplace=True)
```

```python
# Filtras datos de una columna 
df_online_retail_data_filldesc[df_online_retail_data_filldesc['Description'] == 'Descripción desconocida']
```

Completar los valores nulos en 'CustomerID' relacionandolo con el InvoivceNo

```python
# Duplicarl el DataFrame df_online_retail_data_filldesc como un DataFrame temporal
df_online_retail_data_trans = df_online_retail_data_filldesc.copy()
df_online_retail_data_trans.info()
```

```python
# Rellenar los valores nulos en la columna 'CustomerID' con valores de la columna 'InvoiceNo'
df_online_retail_data_trans["CustomerID"] = df_online_retail_data_trans.groupby("InvoiceNo")["CustomerID"].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 12345))
```
